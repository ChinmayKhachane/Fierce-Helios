```python
def getEarnings(Symbol, API_KEY):

  try:
    response = requests.get(f'https://financialmodelingprep.com/api/v3/historical/earning_calendar/{Symbol}?apikey={API_KEY}')
    data = response.json()

    df = pd.DataFrame(data)

    df['date'] = pd.to_datetime(df['date'], errors='coerce')

    df = df.loc[df['date'].dt.year >= 2018]
    df = df.dropna()
    return df

  except:
    print(f'{Symbol} Error')
    return None









```

```python
def getPrice(row):

  startDate = row["startDate"].strftime("%Y-%m-%d")
  endDate = row["endDate"].strftime("%Y-%m-%d")
  symbol = row["symbol"]



  response = requests.get(f'https://financialmodelingprep.com/api/v3/historical-price-full/{symbol}?from={startDate}&to={endDate}&apikey=qlSr2TXDXZi2cX7RTLWhbSDI3R5ZB09z')
  data = response.json()
  time.sleep(0.3)
  old, recent = filterFirstAndLast(data)
  return old, recent



```

```python
def filterFirstAndLast(data):
  try:
    data = data["historical"]
  except:
    return None, None

  recent = data[0]["close"]
  old = data[-1]["close"]
  return old, recent
```

Price Function

```python
def setPriceBeforeAfter(df):
  df = df.drop(columns=['updatedFromDate'])
  df["date"] = pd.to_datetime(df["date"])
  df["startDate"] = df["date"] - pd.Timedelta(days=3)
  df["endDate"] = df["date"] + pd.Timedelta(days=4)

  df["Price3DaysBefore"], df["Price3DaysAfter"] = zip(*df.apply(getPrice, axis=1))
  return df
```
